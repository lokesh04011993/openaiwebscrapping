```python
import os
import requests
from dotenv import load_dotenv
from bs4 import BeautifulSoup
from IPython.display import Markdown, display
from openai import OpenAI
```

```python
load_dotenv()
apikey=os.getenv("key")
```

```python
#check the key
if not apikey:
  print("no api key")
elif apikey[:8]!="sk-proj-":
  print("api key is found..it will start with sk-proj- check api key")
elif apikey.strip()!=apikey:
  print("api key is found ,remove the space if it is found")
else:
  print("api key is found and it is correct")
```

```python
from openai import OpenAI
openai=OpenAI(api_key=apikey)
```

```python
class website:
  url:str
  title:str
  text:str
  def __init__(self,url):
    self.url=url
    response=requests.get(url)
    soup=BeautifulSoup(response.text,"html.parser")
    self.title=soup.title.string if soup.title else "no title found"
    for irr in soup.find_all("script","style","img","input"):
      irr.decompose()
    self.text=soup.body.get_text()
    print(self.text)
```

```python
#lets take any one website
#zz=input("enter the website = ")
zz="https://mrcet.com/"
aa=website(zz)
print(aa.title)
print(aa.text)
```

```python
systemprompt="Tou are an assistant that analyses the content of a website \
and provide a short summary,ignore text that are not part of the website \
Respond in Markdown"
```

```python
#user prompt
def userprompt(website):
  users=f"you are lloinf for a title of the website{website.title}"
  users+="The conent of the website is"
  users+=website.text
  return users
```

```python
def message(userprompt):
  return [
    {"role":"system","content":systemprompt},
    {"role":"user","content":userprompt}
  ]
```

```python
print(message(aa),sep="\n")
```

```python
message(aa.text)
```

```python
#calling open api key
def summary(url):
  w=website(url)
  response=openai.chat.completions.create(
      model="gpt-3.5-turbo",
      messages=message(userprompt(w))
  )
  return response.choices[0].message.content
```

```python
summary(zz)
```
